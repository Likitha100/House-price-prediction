```python
#data cleaning
#handle missing value
df.fillna(df.select_dtypes(include=[np.number]).mean(), inplace=True)
```

```python
#Removing Duplicates
df.drop_duplicates(inplace=True)
```

```python
#DATA ANALYSIS
# 1. Total Property Value (Total Sales)
total_price = np.sum(df['price'])
print("Total Property Value (Sales): $", total_price)
```

```python
# 2. Average Lot Size
average_lot_size = np.mean(df['lotsize'])
print("Average Lot Size:", round(average_lot_size, 2), "sq ft")
```

```python
# 3.Total Number of Houses with Air Conditioning
total_airco = np.sum(df['airco'] == 'yes')
print("Houses with Air Conditioning:", total_airco)
```

```python
#DATA VISUALIZATIONS
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Load the dataset
df = pd.read_csv("Housing.csv")
```

```python
#1.Histogram of House Prices
plt.figure(figsize=(8, 5))
sns.histplot(df['price'], bins=30, kde=True, color='skyblue')
plt.title('Distribution of House Prices')
plt.xlabel('Price')
plt.ylabel('Count')
plt.tight_layout()
plt.show()
```

```python
#2.Bar Plot of Air Conditioning Availability
plt.figure(figsize=(6, 5))
sns.countplot(x='airco', data=df, palette='Set2')
plt.title('Air Conditioning Presence')
plt.xlabel('Air Conditioning')
plt.ylabel('Number of Houses')
plt.tight_layout()
plt.show()
```

```python
#3.Scatter Plot of Lot Size vs Price
plt.figure(figsize=(8, 5))
sns.scatterplot(x='lotsize', y='price', data=df, color='orange')
plt.title('Lot Size vs Price')
plt.xlabel('Lot Size')
plt.ylabel('Price')
plt.tight_layout()
plt.show()
```
